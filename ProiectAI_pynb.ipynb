```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Inițializăm LabelEncoder
le = LabelEncoder()

# Selectăm rândurile fără valori 0 pentru antrenare
train_data = df[df['Breed'] != -1]

# Codificăm valorile din coloana 'Breed'
y_train = le.fit_transform(train_data['Breed'])

# Eliminăm coloana 'Breed' pentru a pregăti setul de antrenare
X_train = train_data.drop(columns=['Breed'])

# Selectăm rândurile cu valori 0 în coloana 'Breed'
missing_data = df[df['Breed'] == -1]

# Pregătim datele lipsă pentru predicție
X_missing = missing_data.drop(columns=['Breed'])

# Inițializăm și antrenăm RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Facem predicții pentru rândurile lipsă
predicted_values_numeric = rf.predict(X_missing)

# Decodificăm predicțiile înapoi la valorile inițiale
predicted_values = le.inverse_transform(predicted_values_numeric)

# Înlocuim valorile 0 din 'Breed' cu predicțiile realizate
df.loc[df['Breed'] == -1, 'Breed'] = predicted_values

# Afișăm DataFrame-ul pentru a verifica rezultatele
print(df)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.utils import shuffle


class Network:

    def __init__(self, sizes):
        self.num_layers = len(sizes)
        self.sizes = sizes

        self.biases = [np.random.randn(y, 1) for y in self.sizes[1:]]
        self.weights = [np.random.randn(y, x) for x, y in zip(self.sizes[:-1], self.sizes[1:])]

    def feedforward(self, a):
      for i, (b, w) in enumerate(zip(self.biases, self.weights)):
          z = np.dot(w, a) + b
          if i == len(self.biases) - 1:
              a = softmax(z)
          else:
              a = sigmoid(z)
      return a



    def SGD(self, training_data, epochs, mini_batch_size, eta, test_data=None):

        n = len(training_data)
        for j in range(epochs):
            shuffle(training_data)
            mini_batches = [
                training_data[k:k + mini_batch_size]
                for k in range(0, n, mini_batch_size)]
            for mini_batch in mini_batches:
                self.update_mini_batch(mini_batch, eta)
            if test_data:

                accuracy_train = self.accuracy(training_data) / len(training_data) * 100
                print(f"Epoch {j}: Acuratetea la antrenare: {accuracy_train:.2f}%")


                accuracy_test = self.accuracy(test_data) / len(test_data) * 100
                print(f"Epoch {j}: Acuratetea la testare: {accuracy_test:.2f}%")
                print("--------------------------------------------------------")


    def update_mini_batch(self, mini_batch, eta):

      delta_b = [np.zeros(b.shape) for b in self.biases]
      delta_w = [np.zeros(w.shape) for w in self.weights]


      for x, y in mini_batch:
          grad_b, grad_w = self.backprop(x, y)

          for i in range(len(delta_b)):
              delta_b[i] += grad_b[i]

          for i in range(len(delta_w)):
              delta_w[i] += grad_w[i]

      for i in range(len(self.weights)):
          self.weights[i] -= (eta / len(mini_batch)) * delta_w[i]

      for i in range(len(self.biases)):
          self.biases[i] -= (eta / len(mini_batch)) * delta_b[i]

    def backprop(self, x, y):

        delta_b = [np.zeros(b.shape) for b in self.biases]
        delta_w = [np.zeros(w.shape) for w in self.weights]

        activation = x
        activations = [x]
        zs = []

        #feedforward
        for b, w in zip(self.biases, self.weights):
            z = np.dot(w, activation) + b
            zs.append(z)
            if w is self.weights[-1]:
                activation = softmax(z)
            else:
                activation = sigmoid(z)
            activations.append(activation)

        # backward propagation
        delta = activations[-1] - y
        delta_b[-1] = delta
        delta_w[-1] = np.dot(delta, activations[-2].transpose())

        for l in range(2, self.num_layers):
            z = zs[-l]
            sp = sigmoid_prime(z)
            delta = np.dot(self.weights[-l + 1].transpose(), delta) * sp
            delta_b[-l] = delta
            delta_w[-l] = np.dot(delta, activations[-l - 1].transpose())
        return (delta_b, delta_w)


    def accuracy(self, data):
      results = [(np.argmax(self.feedforward(x)), np.argmax(y)) for (x, y) in data]
      #for predicted, actual in results:
         # print(f"Valoare prezisa: {predicted}, Valoare reala: {actual}")
      return sum(int(x == y) for (x, y) in results)

    def cost_derivative(self, output_activations, y):
        return (output_activations - y)

    def cross_entropy_loss(self,a,y):
         return np.sum(np.nan_to_num(-y*np.log(a)-(1-y)*np.log(1-a)))

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_prime(z):
    return sigmoid(z) * (1 - sigmoid(z))

def softmax(z):
    exp_z = np.exp(z - np.max(z))
    return exp_z / np.sum(exp_z, axis=0, keepdims=True)

def main():

    X = df.drop(columns=['Breed']).values
    y = df['Breed'].values


    encoder = OneHotEncoder(sparse_output=False)
    y = encoder.fit_transform(y.reshape(-1, 1))



    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


    training_data = [(x.reshape(-1, 1), y.reshape(-1, 1)) for x, y in zip(X_train, y_train)]
    test_data = [(x.reshape(-1, 1), y.reshape(-1, 1)) for x, y in zip(X_test, y_test)]


    input_size = X.shape[1]
    output_size = y.shape[1]

    net = Network([input_size,25,25,output_size])
    net.SGD(training_data, epochs=30, mini_batch_size=32, eta=0.1, test_data=test_data)

if __name__ == "__main__":
    main()
```

```python
import sys
import nltk
nltk.download('stopwords')

from langdetect import detect

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

#Citirea de la tastatura sau din fisier
if len(sys.argv)>1:
  file_name=sys.argv[1]
  with open(file_name,'r',encoding='utf-8') as file:
    text=file.read()
else:
  text=input('Introduceti textul aici:')


#Recunoasterea limbii
language=detect(text)
print('Limba detectata:',language)


#Informatii stilometrice
tokens=word_tokenize(text)

stop_words = stopwords.words('english' if language not in stopwords.fileids() else language)
filtered_words = [word for word in tokens if word.lower() not in stop_words]
print("Cuvinte filtrate:", filtered_words)

freq_distribution=FreqDist(filtered_words)
print("Frecventa cuvintelor:",freq_distribution.most_common())

if len(filtered_words) > 0:
    lexical_diversity = len(set(filtered_words)) / len(filtered_words)
    print("Diversitate lexicala:", lexical_diversity)
else:
    print("Nu exista cuvinte pentru calculul diversitatii lexicale.")


if len(filtered_words) > 0:
    average_word_length = sum(len(word) for word in filtered_words) / len(filtered_words)
    print("Lungimea medie a cuvintelor:", average_word_length)
else:
    print("Nu exista cuvinte pentru calculul lungimii medii.")

```
